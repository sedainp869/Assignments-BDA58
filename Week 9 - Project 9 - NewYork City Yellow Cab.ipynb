```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('NY_YellowTaxi_Trip_data.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
# Drop rows with any missing values 
df = df.dropna()

# Convert date columns to datetime format if not already done
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Calculate trip duration in minutes
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# Filter out trips with zero or negative duration or distance (if any)
df = df[(df['trip_duration'] > 0) & (df['trip_distance'] > 0)]

# Check for duplicates
df = df.drop_duplicates()
df
```

```python
df.isnull().sum()
```

```python
# Plot trip duration distribution
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['trip_duration'], bins=50, kde=True)
plt.title('Distribution of Trip Durations (minutes)')
plt.xlabel('Trip Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot fare amount distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['fare_amount'], bins=50, kde=True)
plt.title('Distribution of Fare Amounts')
plt.xlabel('Fare Amount (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Extract hour from pickup datetime
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour

# Plot peak hours
plt.figure(figsize=(12, 6))
sns.countplot(x='pickup_hour', data=df, palette='viridis')
plt.title('Trips by Pickup Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Trips')
plt.show()
```

```python
# Create a new column for routes
df['route'] = df['pickup_longitude'].astype(str) + ',' + df['pickup_latitude'].astype(str) + \
              " to " + df['dropoff_longitude'].astype(str) + ',' + df['dropoff_latitude'].astype(str)

# Plot top 10 popular routes
top_routes = df['route'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(y=top_routes.index, x=top_routes.values, palette="coolwarm")
plt.title('Top 10 Popular Routes')
plt.xlabel('Number of Trips')
plt.ylabel('Route')
plt.show()
```

```python
# Create histogram of tip_amount
plt.figure(figsize=(12,6))
ax = sns.histplot(df['tip_amount'], bins=range(0,21,1))
ax.set_xticks(range(0,21,2))
ax.set_xticklabels(range(0,21,2))
plt.title('Tip amount histogram');
```

```python

# Ensure 'tpep_pickup_datetime' is in datetime format
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce')

# Extracting day of the week from 'tpep_pickup_datetime'
df['day_of_week'] = df['tpep_pickup_datetime'].dt.day_name()

# Plotting taxi demand by day of the week
plt.figure(figsize=(10, 6))
sns.countplot(x='day_of_week', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Taxi Demand by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Trips')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Make sure 'trip_duration_minutes' exists in your DataFrame.
# Calculate 'trip_duration_minutes' if not already present:
if 'trip_duration_minutes' not in df.columns:
    df['trip_duration_minutes'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# Use Spearman correlation and plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['fare_amount', 'total_amount', 'trip_distance', 'trip_duration_minutes']].corr(method='spearman'), 
            annot=True, cmap='coolwarm', square=True)
plt.title('Spearman Correlation between Trip Metrics')
plt.show()
```
